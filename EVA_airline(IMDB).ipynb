```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

movie_name = '.text-regular'
requests_url = 'https://www.evaair.com/zh-tw/fly-prepare/flying-with-eva/inflight-entertainment-service/movies/'
response = requests.get(requests_url)
soup = BeautifulSoup(response.text)
# print(type(soup))
# M_name = [for e in soup.select(movie_name).text]
# M_name
# result = [e.text for e in soup.select(movie_name)]
result= []

for e in soup.select(movie_name):
  result.append(e.text)



# 取得評分
def wr(requests_url):
  response = requests.get(requests_url)
  soup = BeautifulSoup(response.text)
  M_name = soup.find('h1').text
  M_score = soup.find('span', class_="sc-7ab21ed2-1").text
  M_data = {
      "Movie" : M_name,
      "score" : M_score
  }
  return M_score

# 取得imdb電影連結
def M_name(name):

  data1 = '.result_text > a'
  query_string_parameters = {
      'q': name,
      's': 'tt',
      'ttype': 'ft',
      'ref_': 'fn_ft'
  }
  requests_url = 'https://www.imdb.com/find'
  response = requests.get(requests_url,params = query_string_parameters)
  soup = BeautifulSoup(response.text)
  result = [e.get('href') for e in soup.select(data1)]
  URL = 'https://www.imdb.com/'+result[0]
  return URL


Movie_data = pd.DataFrame()
Name = []
Score = []
Web = []
# for r in result:
#   try:
#     Movie_data  = M_name(r)
#     print(Movie_data)
#   except:
#     print(r,"未取得資料")
  
for r in result:
  try:
    Name.append(r)
    Web.append(M_name(r))
    Score.append(wr(M_name(r)))
  except:
    Web.append('Na')
    Score.append('Na')
Movie_data['電影名稱'] = Name
Movie_data['Score'] = Score
Movie_data['IMDB'] = Web
Movie_data
```
